```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("churn.db")
print("Connected to churn.db database successfully.")
```

```python
query1 = """
SELECT 
    city,
    COUNT(*) AS total_customers,
    SUM(churn) AS churned_customers,
    ROUND(SUM(churn) * 100.0 / COUNT(*), 2) AS churn_rate_pct
FROM clean_master
GROUP BY city
ORDER BY churn_rate_pct DESC
"""

result1 = pd.read_sql_query(query1, conn)
print("Churn Rate by City:")
print(result1)
```

```python
query2 = """
SELECT 
    primary_product,
    COUNT(*) AS total_customers,
    SUM(churn) AS churned_customers,
    ROUND(SUM(churn) * 100.0 / COUNT(*), 2) AS churn_rate_pct
FROM clean_master
GROUP BY primary_product
ORDER BY churn_rate_pct DESC
"""

result2 = pd.read_sql_query(query2, conn)
print("Churn Rate by Product:")
print(result2)
```

```python
query3 = """
SELECT 
    CASE WHEN churn = 1 THEN 'Churned' ELSE 'Active' END AS customer_status,
    COUNT(*) AS total_customers,
    ROUND(AVG(total_transactions), 1) AS avg_transactions,
    ROUND(AVG(total_transaction_amount), 2) AS avg_transaction_amount
FROM clean_master
GROUP BY churn
"""

result3 = pd.read_sql_query(query3, conn)
print("Transaction Behaviour — Churned vs Active:")
print(result3)
```

```python
query4 = """
SELECT 
    customer_id,
    city,
    primary_product,
    active_lending_relationships,
    total_transactions,
    high_risk_tickets,
    churn
FROM clean_master
WHERE churn = 1
AND high_risk_tickets >= 1
AND active_lending_relationships >= 4
ORDER BY active_lending_relationships DESC, high_risk_tickets DESC
LIMIT 10
"""

result4 = pd.read_sql_query(query4, conn)
print("Top 10 Highest Risk Customers:")
print(result4)
```

```python
conn.close()
print("Database connection closed.")
```
